```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import json
import seaborn as sns
import os
import glob
```

```python
extension = 'csv'
all_filenames = [i for i in glob.glob('*.{}'.format(extension))]
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in all_filenames ])
```

```python
combined_csv.to_csv( "combined_csv.csv", index=False, encoding='utf-8-sig')
```

```python
df = pd.read_csv('combined_csv.csv',names=['palavra','dias','prob'], header=0)
```

```python
df.shape
```

```python
df
```

```python
result = df.pivot_table(index='palavra', columns='dias', values='prob', aggfunc='sum')
```

```python
result
```

```python
fig, ax = plt.subplots(figsize=(17,22))

title = "Twitter heatmap"

plt.title(title,fontsize=18)
ttl = ax.title
ttl.set_position([0.5,1.05])
sns.heatmap(result,fmt="",cmap='Purples',ax=ax)

plt.savefig('heatmap_completo.png', format='png')
plt.show()
```
